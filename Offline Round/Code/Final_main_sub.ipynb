# Date

```python
import dateutil.parser as dparser
import datetime
def find_date(val):
    try:
        print('dparser',dparser.parse(val,fuzzy=True))
        t = dparser.parse(val,fuzzy=True)
        print(t.strftime('%Y%m%d'))
        return t.strftime('%Y%m%d')
    except:
        return 'NotFound'

punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''     
def remove_punc(v):
    vv=""
    for x in v:
        if x not in punctuations:
            vv=vv+x
    return vv
```

```python
import glob
from bs4 import BeautifulSoup
i=0
paa=[]
pbb=[]
month_list=['january','feb','march','april','may','june','july','august','september','october','november','december']
# print(glob.glob("ML/training_data/data/*.xml"))
for k in sub['File_Name'].values:
    infile = open("ML/private_test_data/private_test_data/"+k+".xml","r",encoding="utf8")
    
    contents = infile.read()
    soup = BeautifulSoup(contents,'xml')
    titles = soup.find_all('text')
    print('---------------',k,'------------')
    fa=True
    fb=True
    for title in titles:
        pA=[m for m in party_A if remove_punc(m.lower()) in remove_punc(title.get_text().lower())]
        pB=[n for n in party_B if remove_punc(n.lower()) in remove_punc(title.get_text().lower())]
        if pA!=[] and len(paa)==i:
            print("partyA",pA[0])
            paa.append(pA[0])
            fa=False
        if pB!=[] and len(pbb)==i:
            print("partyB",pB[0])
            pbb.append(pB[0])
            fb=False
    if fa==True or fb==True:
        paa.append('NotFound')
        pbb.append('NotFound')
    i+=1
    
```
