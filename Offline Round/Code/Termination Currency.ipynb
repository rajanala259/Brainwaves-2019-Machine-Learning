```python
curr_df=pd.DataFrame({'lab_curr':lab['TC_currency'],'txt':curr})
curr_df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
def to_words(raw_text):
    raw_text=raw_text.strip()
    no_coms=re.sub(r'\.com','',raw_text)
    no_urls=re.sub('https?://www','',no_coms)
    no_urls1=re.sub('https?://','',no_urls)
    try:
        no_encoding=no_urls1.decode("utf-8-sig").replace(u"\ufffd", "?")
    except:
        no_encoding = no_urls1
    letters_only = re.sub("[^a-zA-Z]", " ",no_encoding) 
    words = letters_only.lower().split()                             
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops] 
    return( " ".join( meaningful_words ))
```

```python
curr_df['clean_txt']=curr_df['txt'].apply(to_words)
```

```python
curr_df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_val,y_train,y_val = train_test_split(curr_df['clean_txt'],curr_df['lab_curr'],
                                                 test_size=0.25,random_state = 1994)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec=TfidfVectorizer(ngram_range=(1,3))
xtvec=vec.fit_transform(X_train)
xtstvec=vec.transform(X_val)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=1994,n_estimators=200)
rf.fit(xtvec,y_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score
print(accuracy_score(y_val,rf.predict(xtstvec)))
print(f1_score(y_val,rf.predict(xtstvec),average='weighted'))
```
