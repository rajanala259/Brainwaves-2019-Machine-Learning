```python
import pandas as pd
import numpy as np
```

```python
lab=pd.read_csv('ML/train_labels.csv')
sub=pd.read_csv('ML/submission_format6c29943.csv')
```

# Train data

```python
from bs4 import BeautifulSoup

thrs_peramt=[]
i=0

for k in lab['File_Name'].values:
    infile = open("ML/training_data/data/"+k+".xml","r",encoding="utf8")
    contents = infile.read()
    f=True
    soup = BeautifulSoup(contents,'xml')
    titles = soup.find_all('text')
    print('---------------',k,'------------')
    
    for title in titles:
        if title.get_text().find('Threshold Amount')!=-1:
            print('>>',title.get_text())
            thrs_peramt.append(title.get_text())
            f=False
            break
    if f==True:
        thrs_peramt.append('NotFound')
        
    i+=1
        
```

```python
import re
import nltk
from nltk.corpus import stopwords
def to_numbers(raw_text):
    raw_text=raw_text.strip()
    no_coms=re.sub(r'\.com','',raw_text)
    no_urls=re.sub('https?://www','',no_coms)
    no_urls1=re.sub('https?://','',no_urls)
    try:
        no_encoding=no_urls1.decode("utf-8-sig").replace(u"\ufffd", "?")
    except:
        no_encoding = no_urls1
#     letters_only = re.sub("[^0-9]", "",no_encoding) 
#     words = letters_only.lower().split()                             
#     stops = set(stopwords.words("english"))                  
#     meaningful_words = [w for w in no_encoding if not w in stops] 
    punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    meaningful_words = [w for w in no_encoding if not w in punctuations] 
    return( "".join( meaningful_words ))
```

```python
thrshldamtperdf=pd.DataFrame({'txt':thrs_peramt,'thrs_peramt':lab['Cross_default_percentage/amount']})
thrshldamtperdf.head()
```

```python
thrshldamtperdf['clean_txt']=thrshldamtperdf['txt'].apply(to_numbers)
```
